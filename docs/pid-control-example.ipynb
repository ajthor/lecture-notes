```python
%%markdown
## What is a PID Controller?

A Proportional-Integral-Derivative (PID) controller is a widely used feedback control loop mechanism in industrial control systems. A PID controller calculates an "error" value as the difference between a measured process variable and a desired setpoint. The controller attempts to minimize the error by adjusting a control variable.

The PID controller's output is a weighted sum of three terms:

1.  **Proportional (P) term:** Accounts for the present value of the error.
2.  **Integral (I) term:** Accounts for the accumulation of past errors.
3.  **Derivative (D) term:** Accounts for the prediction of future errors based on the current rate of change.

The goal of a PID controller is to regulate a system's output to a desired setpoint by continuously adjusting a control input based on the feedback error.
```

**Reasoning**:
Create a markdown cell explaining the Proportional (P) term, its effect on the system, and the concept of proportional gain (Kp).



```python
%%markdown
## The Proportional (P) Term

The proportional term of a PID controller is directly proportional to the current error. The larger the error, the larger the proportional output will be. This term provides an immediate response to the error.

The contribution of the proportional term to the control output is given by:

$P_{out} = K_p \times e(t)$

where:
*   $P_{out}$ is the proportional output.
*   $K_p$ is the **proportional gain**, a tuning parameter.
*   $e(t)$ is the current error at time $t$.

A higher proportional gain ($K_p$) results in a larger change in the output for a given error. While a higher $K_p$ can reduce the rise time and decrease the steady-state error, it can also lead to instability, oscillations, and even overshoot in the system response. A low $K_p$ will result in a slower response and may not be able to eliminate the steady-state error entirely.
```

**Reasoning**:
Create a markdown cell explaining the Integral (I) term, its effect on the system (especially regarding steady-state error), and the concept of integral gain (Ki).



```python
%%markdown
## The Integral (I) Term

The integral term of a PID controller accounts for the accumulation of past errors over time. This term is essential for eliminating steady-state error, which is the difference between the desired setpoint and the actual process variable after the system has settled. If there is a persistent error, the integral term will grow over time, adding to the control output until the error is reduced to zero.

The contribution of the integral term to the control output is given by:

$I_{out} = K_i \times \int_{0}^{t} e(\tau) d\tau$

where:
*   $I_{out}$ is the integral output.
*   $K_i$ is the **integral gain**, a tuning parameter.
*   $\int_{0}^{t} e(\tau) d\tau$ is the integral of the error over time from the start (0) to the current time ($t$).

A higher integral gain ($K_i$) makes the controller more aggressive in eliminating steady-state error, but it can also lead to overshoot and oscillations, and potentially windup if not handled properly. A low $K_i$ will result in a slower elimination of steady-state error.
```

**Reasoning**:
Create a markdown cell explaining the Derivative (D) term, its effect on the system (especially regarding damping and overshoot), and the concept of derivative gain (Kd).



```python
%%markdown
## The Derivative (D) Term

The derivative term of a PID controller anticipates future errors by considering the rate of change of the current error. It provides damping to the system, helping to reduce overshoot and improve stability by counteracting rapid changes in the error. This term is particularly useful in systems where quick responses and minimal oscillations are desired.

The contribution of the derivative term to the control output is given by:

$D_{out} = K_d \times \frac{de(t)}{dt}$

where:
*   $D_{out}$ is the derivative output.
*   $K_d$ is the **derivative gain**, a tuning parameter.
*   $\frac{de(t)}{dt}$ is the derivative of the error with respect to time, representing the rate of change of the error.

A higher derivative gain ($K_d$) provides more damping, which can help to reduce overshoot and settling time. However, too high a $K_d$ can make the system overly sensitive to noise in the error signal and can lead to oscillations or instability. A low $K_d$ will result in less damping and the system may exhibit more overshoot.
```

**Reasoning**:
Create a markdown cell explaining the overall PID control equation and how the three terms combine to produce the control output.



```python
%%markdown
## The PID Control Equation

The total output of a PID controller is the sum of the proportional, integral, and derivative terms. This output is the control signal that is sent to the system to adjust the process variable and reduce the error.

The overall PID control output ($u(t)$) is given by the equation:

$u(t) = P_{out} + I_{out} + D_{out}$

Substituting the expressions for each term:

$u(t) = K_p \times e(t) + K_i \times \int_{0}^{t} e(\tau) d\tau + K_d \times \frac{de(t)}{dt}$

where:
*   $u(t)$ is the control output at time $t$.
*   $K_p$, $K_i$, and $K_d$ are the proportional, integral, and derivative gains, respectively.
*   $e(t)$ is the current error at time $t$.
*   $\int_{0}^{t} e(\tau) d\tau$ is the integral of the error over time.
*   $\frac{de(t)}{dt}$ is the derivative of the error with respect to time.

By adjusting the values of the gains ($K_p$, $K_i$, and $K_d$), the PID controller can be tuned to achieve desired performance characteristics such as rise time, overshoot, settling time, and steady-state error for a specific system.
```

```python
%%markdown
## Tuning PID Parameters

Tuning a PID controller involves finding the optimal values for the proportional gain ($K_p$), integral gain ($K_i$), and derivative gain ($K_d$) to achieve the desired system response. The goal is typically to minimize the error, reduce overshoot, decrease settling time, and maintain stability.

Several methods exist for tuning PID controllers, ranging from empirical trial-and-error to more systematic approaches:

*   **Trial and Error:** Adjusting the gains manually based on observing the system's response. This can be time-consuming and requires experience.
*   **Ziegler-Nichols Method:** A well-known empirical tuning method that involves finding the ultimate gain and period of oscillation of the system under proportional control only.
*   **Software-Based Tuning:** Using specialized software that can analyze the system's characteristics and suggest optimal PID parameters.
*   **Model-Based Tuning:** Developing a mathematical model of the system and using it to calculate the optimal PID parameters.

The best tuning method depends on the specific system, its complexity, and the desired performance criteria. Often, an iterative process of tuning and testing is required to achieve the desired control performance.
```

## Implement mass-spring-damper model

### Subtask:
Write Python code using numpy and scipy to simulate a mass-spring-damper system.


**Reasoning**:
Define the differential equation for the mass-spring-damper system and use scipy.integrate.odeint to simulate the system's response over time based on the provided parameters and initial conditions.



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the differential equation for the mass-spring-damper system
def mass_spring_damper(y, t, m, k, c):
    """
    Defines the differential equation for a mass-spring-damper system.

    Args:
        y: A list containing the current state [position, velocity].
        t: The current time.
        m: Mass.
        k: Spring constant.
        c: Damping coefficient.

    Returns:
        A list containing the derivatives of position and velocity [velocity, acceleration].
    """
    x, v = y  # Unpack the state: position and velocity
    dxdt = v  # Derivative of position is velocity
    dvdt = (-k * x - c * v) / m  # Derivative of velocity is acceleration (from F=ma)
    return [dxdt, dvdt]

# Define system parameters
m = 1.0  # Mass (kg)
k = 10.0  # Spring constant (N/m)
c = 1.0  # Damping coefficient (Ns/m)

# Define initial conditions
initial_position = 1.0  # Initial position (m)
initial_velocity = 0.0  # Initial velocity (m/s)
initial_state = [initial_position, initial_velocity]

# Define time span for simulation
t_start = 0.0  # Start time (s)
t_end = 10.0  # End time (s)
num_points = 200  # Number of time points
time_points = np.linspace(t_start, t_end, num_points)

# Simulate the system using odeint
solution = odeint(mass_spring_damper, initial_state, time_points, args=(m, k, c))

# Extract position and velocity from the solution
position = solution[:, 0]
velocity = solution[:, 1]

# Display the first few time points, positions, and velocities
print("Time (s) | Position (m) | Velocity (m/s)")
print("-" * 40)
for i in range(min(10, num_points)):
    print(f"{time_points[i]:<8.2f} | {position[i]:<12.4f} | {velocity[i]:<12.4f}")
```

## Implement pid controller

### Subtask:
Write Python code to implement a PID controller.


**Reasoning**:
Define a function to implement the PID controller as described in the instructions.



```python
def pid_controller(error, previous_error, integral_error, Kp, Ki, Kd):
    """
    Calculates the output of a PID controller.

    Args:
        error: The current error (setpoint - process variable).
        previous_error: The error from the previous time step.
        integral_error: The accumulated integral of the error.
        Kp: Proportional gain.
        Ki: Integral gain.
        Kd: Derivative gain.

    Returns:
        The PID controller output.
    """
    # Proportional term
    proportional_term = Kp * error

    # Integral term
    integral_term = Ki * integral_error

    # Derivative term
    derivative_term = Kd * (error - previous_error)

    # Total PID output
    output = proportional_term + integral_term + derivative_term

    return output
```

## Combine model and controller

### Subtask:
Integrate the PID controller with the mass-spring-damper model.


**Reasoning**:
Modify the mass_spring_damper function to accept a control input, and then integrate the PID controller into a simulation loop to control the mass-spring-damper system. This involves calculating the error, integral error, and PID output at each time step, and using the PID output as an additional force in the mass-spring-damper dynamics.



```python
# Modify the mass_spring_damper function to accept control input
def mass_spring_damper_controlled(y, t, m, k, c, control_input):
    """
    Defines the differential equation for a mass-spring-damper system with a control input.

    Args:
        y: A list containing the current state [position, velocity].
        t: The current time.
        m: Mass.
        k: Spring constant.
        c: Damping coefficient.
        control_input: The control force from the PID controller.

    Returns:
        A list containing the derivatives of position and velocity [velocity, acceleration].
    """
    x, v = y  # Unpack the state: position and velocity
    dxdt = v  # Derivative of position is velocity
    # Add the control input to the force equation
    dvdt = (-k * x - c * v + control_input) / m  # Derivative of velocity is acceleration (from F=ma)
    return [dxdt, dvdt]

# Define PID gains (these will be made interactive later)
Kp = 50.0
Ki = 5.0
Kd = 10.0

# Define setpoint
setpoint = 0.5  # Desired position (m)

# Initialize variables for the simulation loop
initial_position = 1.0  # Initial position (m)
initial_velocity = 0.0  # Initial velocity (m/s)
initial_state = [initial_position, initial_velocity]

t_start = 0.0  # Start time (s)
t_end = 10.0  # End time (s)
num_points = 200  # Number of time points
time_points = np.linspace(t_start, t_end, num_points)
dt = time_points[1] - time_points[0] # Time step

# Initialize storage for results
positions = []
velocities = []
errors = []
control_outputs = []

# Initialize PID variables
previous_error = 0.0
integral_error = 0.0
current_state = initial_state

# Simulation loop
for i in range(num_points):
    current_time = time_points[i]
    current_position = current_state[0]

    # Calculate current error
    error = setpoint - current_position

    # Update integral error (using rectangular approximation)
    integral_error += error * dt

    # Calculate PID controller output
    control_output = pid_controller(error, previous_error, integral_error, Kp, Ki, Kd)

    # Store current values
    positions.append(current_position)
    velocities.append(current_state[1])
    errors.append(error)
    control_outputs.append(control_output)

    # Update the mass-spring-damper system state using odeint for one time step
    # We need to pass the control_input as an argument to the function
    next_state = odeint(mass_spring_damper_controlled, current_state, [current_time, current_time + dt], args=(m, k, c, control_output))

    # Update the current state for the next iteration
    current_state = next_state[1]


    # Update previous error for the next iteration
    previous_error = error

# Convert lists to numpy arrays for easier plotting/analysis
positions = np.array(positions)
velocities = np.array(velocities)
errors = np.array(errors)
control_outputs = np.array(control_outputs)

# Display the first few time points, positions, errors, and control outputs
print("Time (s) | Position (m) | Error (m) | Control Output (N)")
print("-" * 50)
for i in range(min(10, num_points)):
    print(f"{time_points[i]:<8.2f} | {positions[i]:<12.4f} | {errors[i]:<9.4f} | {control_outputs[i]:<16.4f}")
```

## Create interactive widget

Use `ipywidgets` to create an interactive widget for tuning PID parameters.

Import necessary modules from ipywidgets and matplotlib. Define a function that simulates the system with given PID parameters and plots the results. Create interactive widgets for PID parameters and link them to the simulation function. Display the widget and the initial plot.



```python
from ipywidgets import interactive, FloatSlider
import matplotlib.pyplot as plt

# Define a function to run the simulation and plot the results
def run_simulation(Kp, Ki, Kd):
    """
    Runs the mass-spring-damper simulation with a PID controller
    and plots the position over time.

    Args:
        Kp: Proportional gain.
        Ki: Integral gain.
        Kd: Derivative gain.
    """
    # Initialize variables for the simulation loop
    initial_position = 1.0  # Initial position (m)
    initial_velocity = 0.0  # Initial velocity (m/s)
    initial_state = [initial_position, initial_velocity]

    t_start = 0.0  # Start time (s)
    t_end = 10.0  # End time (s)
    num_points = 200  # Number of time points
    time_points = np.linspace(t_start, t_end, num_points)
    dt = time_points[1] - time_points[0] # Time step

    # Define setpoint
    setpoint = 0.5  # Desired position (m)

    # Initialize storage for results
    positions = []
    velocities = []
    errors = []
    control_outputs = []

    # Initialize PID variables
    previous_error = 0.0
    integral_error = 0.0
    current_state = initial_state

    # Simulation loop
    for i in range(num_points):
        current_time = time_points[i]
        current_position = current_state[0]

        # Calculate current error
        error = setpoint - current_position

        # Update integral error (using rectangular approximation)
        integral_error += error * dt

        # Calculate PID controller output
        control_output = pid_controller(error, previous_error, integral_error, Kp, Ki, Kd)

        # Store current values
        positions.append(current_position)
        velocities.append(current_state[1])
        errors.append(error)
        control_outputs.append(control_output)

        # Update the mass-spring-damper system state using odeint for one time step
        next_state = odeint(mass_spring_damper_controlled, current_state, [current_time, current_time + dt], args=(m, k, c, control_output))

        # Update the current state for the next iteration
        current_state = next_state[1]

        # Update previous error for the next iteration
        previous_error = error

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.plot(time_points, positions, label='Position')
    plt.axhline(setpoint, color='r', linestyle='--', label='Setpoint')
    plt.xlabel('Time (s)')
    plt.ylabel('Position (m)')
    plt.title('Mass-Spring-Damper System with PID Control')
    plt.legend()
    plt.grid(True)
    plt.show()

# Create interactive widgets for PID parameters
interactive_plot = interactive(run_simulation,
                               Kp=FloatSlider(min=0, max=200, step=1, value=50, description='Kp:'),
                               Ki=FloatSlider(min=0, max=50, step=0.1, value=5, description='Ki:'),
                               Kd=FloatSlider(min=0, max=50, step=0.1, value=10, description='Kd:'))

# Display the interactive widget and the initial plot
display(interactive_plot)
```

## Integrate widget and animation

### Subtask:
Connect the interactive widget to the animation, allowing users to see the effect of changing PID parameters in real-time.


**Reasoning**:
Modify the `run_simulation` function to return the necessary data for the animation and create a new function to generate the animation, then update `run_simulation` to call this animation function and clear previous outputs.



```python
from IPython.display import display, clear_output
import matplotlib.animation as animation

# Define the update function for the animation
def update(frame, line, setpoint_line, time_points, positions, setpoint):
    """
    Updates the data for the animation plot.

    Args:
        frame: The current frame number.
        line: The line object representing the mass position.
        setpoint_line: The line object representing the setpoint.
        time_points: Array of time points.
        positions: Array of mass positions at each time point.
        setpoint: The setpoint value.
    """
    line.set_data(time_points[:frame], positions[:frame])
    setpoint_line.set_data(time_points, np.full_like(time_points, setpoint))
    return line, setpoint_line

# Define the initialization function for the animation
def init():
    """
    Initializes the animation plot.
    """
    line.set_data([], [])
    setpoint_line.set_data([], [])
    return line, setpoint_line

def generate_animation(time_points, positions, setpoint):
    """
    Generates and displays an animation of the mass position over time.

    Args:
        time_points: Array of time points from the simulation.
        positions: Array of mass positions from the simulation.
        setpoint: The setpoint value.
    """
    fig, ax = plt.figure(figsize=(10, 6)), plt.gca()

    line, = ax.plot([], [], lw=2, label='Position')
    setpoint_line, = ax.plot([], [], 'r--', lw=2, label='Setpoint')

    ax.set_xlim(time_points[0], time_points[-1])
    ax.set_ylim(min(min(positions), setpoint) - 0.2, max(max(positions), setpoint) + 0.2)

    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Position (m)')
    ax.set_title('Mass-Spring-Damper System with PID Control Animation')
    ax.legend()
    ax.grid(True)

    ani = animation.FuncAnimation(fig, update, frames=len(time_points),
                                  init_func=init, fargs=(line, setpoint_line, time_points, positions, setpoint),
                                  blit=True)

    plt.close(fig) # Close the initial plot to avoid displaying it twice
    display(ani)


# Modify the run_simulation function to return data and generate animation
def run_simulation(Kp, Ki, Kd):
    """
    Runs the mass-spring-damper simulation with a PID controller,
    returns the results, and generates an animation.

    Args:
        Kp: Proportional gain.
        Ki: Integral gain.
        Kd: Derivative gain.

    Returns:
        Tuple containing:
            - time_points: Array of time points.
            - positions: Array of mass positions.
            - control_outputs: Array of control outputs.
            - setpoint: The setpoint value.
    """
    # Clear previous output
    clear_output(wait=True)

    # Initialize variables for the simulation loop
    initial_position = 1.0  # Initial position (m)
    initial_velocity = 0.0  # Initial velocity (m/s)
    initial_state = [initial_position, initial_velocity]

    t_start = 0.0  # Start time (s)
    t_end = 10.0  # End time (s)
    num_points = 200  # Number of time points
    time_points = np.linspace(t_start, t_end, num_points)
    dt = time_points[1] - time_points[0] # Time step

    # Define setpoint
    setpoint = 0.5  # Desired position (m)

    # Initialize storage for results
    positions = []
    velocities = []
    errors = []
    control_outputs = []

    # Initialize PID variables
    previous_error = 0.0
    integral_error = 0.0
    current_state = initial_state

    # Simulation loop
    for i in range(num_points):
        current_time = time_points[i]
        current_position = current_state[0]

        # Calculate current error
        error = setpoint - current_position

        # Update integral error (using rectangular approximation)
        integral_error += error * dt

        # Calculate PID controller output
        control_output = pid_controller(error, previous_error, integral_error, Kp, Ki, Kd)

        # Store current values
        positions.append(current_position)
        velocities.append(current_state[1])
        errors.append(error)
        control_outputs.append(control_output)

        # Update the mass-spring-damper system state using odeint for one time step
        next_state = odeint(mass_spring_damper_controlled, current_state, [current_time, current_time + dt], args=(m, k, c, control_output))

        # Update the current state for the next iteration
        current_state = next_state[1]

        # Update previous error for the next iteration
        previous_error = error

    # Convert lists to numpy arrays
    positions = np.array(positions)
    control_outputs = np.array(control_outputs)

    # Generate the animation
    generate_animation(time_points, positions, setpoint)

    # Return the results
    return time_points, positions, control_outputs, setpoint

# Create interactive widgets for PID parameters
interactive_plot = interactive(run_simulation,
                               Kp=FloatSlider(min=0, max=200, step=1, value=50, description='Kp:'),
                               Ki=FloatSlider(min=0, max=50, step=0.1, value=5, description='Ki:'),
                               Kd=FloatSlider(min=0, max=50, step=0.1, value=10, description='Kd:'))

# Display the interactive widget
display(interactive_plot)
```
